### Import Dependencies

```python
from langsmith import Client
```

### attach a feedback to a specific trace in langsmith

```python
client = Client()
```

```python
RUN_ID = "019f64a0-eca7-72d1-82e0-ae637bcbe69b"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comment",
    value="This is a long form free text feedback left by TZ",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comment_with_score",
    score=1,
    value="This is a long form free text feedback left by TZ",
    feedback_source_type="api"
)
```
